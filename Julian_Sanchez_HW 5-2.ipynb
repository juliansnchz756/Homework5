```r
Data_Frame <- data.frame (
    ID = c(1,2,3,4,5),
    Name = c('Peter', 'Amy', 'Ryan', 'Gary', 'Michelle'),
    Salary = c(623.30, 515.20, 611.00, 729.00, 843.25))
print(Data_Frame)
```

```r
New_Data_Frame <- cbind(Data_Frame, Department = c('IT', 'Acounting', 'R&D','Finance', 'Economics'))

print(New_Data_Frame)
```

```r
New_Data_Frame[c(1,3,5),c(2,3)]
     
```

```r
x1 <- New_Data_Frame[c(1,4,5),c(2)]
y1 <- New_Data_Frame[c(1,4,5),c(3)]

barplot(y1, names.arg = x1, ylab = 'Salary', xlab = 'Employee', main='Employee Salary')
```

```r
x = c(min(New_Data_Frame[,3]), max(New_Data_Frame[,3]), median(New_Frame[,3]))
mylabels = c("Minimum", "Maximum", "Median")
mycolors = c("red", "blue", "green")

pie(x, label=mylabels, col=mycolors, main="Salaries Stats")
```
